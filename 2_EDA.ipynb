```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
df.sample(4)
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(
    data=df,
    x='Pclass',
    hue='Survived'
)

plt.title('Tasa de Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(
    data=df,
    x='Sex',
    hue='Survived'
)

plt.title('Supervivencia por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])

plt.show()
```

```python
plt.figure(figsize = (12, 8))

# colores para cada una de las clases
color = ['#240750', '#344C64']

sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
plt.legend(title = 'Supervivencia', labels = ['No sobrevivi', 'S sobrevivi'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribucin de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlacin'})  # Etiqueta para la barra de colores
plt.title('Correlacin de variables numricas')
plt.show()
```

Agrupa las edades en categoras (Infante, Nio, Adolescente, Joven Adulto, Adulto, Mayor) y crea una grfica que muestre la tasa de supervivencia por grupo de edad.

```python
rangoEdades = [0, 2, 12, 18, 30, 60, 100]
labels = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']

df['GrupoEdad'] = pd.cut(df['Age'], bins=rangoEdades, labels=labels)

plt.figure(figsize=(12, 8))

sns.barplot(
    data=df,
    x='GrupoEdad',
    y='Survived',
    estimator='mean'
)

plt.title('Tasa de supervivencia por grupo de edad')
plt.xlabel('Grupo de edad')
plt.ylabel('Tasa de supervivencia')

plt.show()
```

Encuentra y muestra los valores mximo y mnimo de la columna de edad.

Crea una grfica que muestre la distribucin de edades, incluyendo lneas verticales que indiquen la media, mediana y moda de las edades.

```python
media = df['Age'].mean()
mediana = df['Age'].median()
moda = df['Age'].mode()[0]

plt.figure(figsize=(12, 8))

sns.histplot(data=df, x='Age', bins=30, kde=True)

plt.axvline(media, linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, linestyle=':', label=f'Moda: {moda:.2f}')

plt.title('Distribucin de edades con media, mediana y moda')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()

plt.show()
```
